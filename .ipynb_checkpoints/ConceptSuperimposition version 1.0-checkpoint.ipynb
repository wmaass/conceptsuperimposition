Explore the correlations between the categorical features
We will use Thiel's U to quantify the relationship between the categorical features.

Theil's U is an uncertainty coefficient that is based on information theory.
It essentially tells us the likelihood of x given y. This probability measure makes Theil's U interpretable.
Moreover, this measure is not symmetric, p(x|y) != p(y|x). This gives us additional information about the relationship between x and y.

```python
#for each categorical variable, dummy code and bind to data
def encode_and_bind(original_dataframe, feature_to_encode):
    print("---------------------")
    print("feat_to_encode: ", feature_to_encode)
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    for i in range(len(dummies.columns)): print("dummy: ", dummies.columns[i])
    print("---------------------")
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res) 

for feature in categorical_feats:
    data = encode_and_bind(data, feature)
```

```python
import xgboost
from xgboost import XGBClassifier
import xgboost
from xgboost import XGBRegressor


def feature_contribution(model_type, data, predictors, outcome, input_concepts):
    X=data[predictors]  # Features
    y=data['action_taken']  # Labels

    # Split dataset into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1) # 70% training and 30% test
    
    if model_type == 'classification':
        model=xgboost.XGBClassifier()
    else:
        model=xgboost.XGBRegressor()

    #Train the model using the training sets y_pred=clf.predict(X_test)
    model.fit(X_train,y_train.values.ravel())
    
    # load the model from disk
    #clf = pickle.load(open(filename, 'rb'))
    result = model.score(X_test, y_test)
    print(result)
    
    y_pred=model.predict(X_test)

#    if model_type == 'classification':        
    # Model Accuracy, how often is the classifier correct?
#        print("Accuracy:", round(metrics.accuracy_score(y_test, y_pred),3))
#        print("Precision: ", round(metrics.precision_score(y_test, y_pred),3))
#        print("Recall: ", round(metrics.recall_score(y_test, y_pred),3))
#        print("F1-score:", round(metrics.f1_score(y_test, y_pred),3))
    
    importances = model.feature_importances_
#    print(pd.DataFrame({'feature': X_train.columns, 'importance': importances}).sort_values('importance', ascending = False).round(2)[:20])
    
    # Create the SHAP explainer instance
    explainer = shap.TreeExplainer(model, X_test)
    #explainer = shap.explainers.GPUTree(model, X_train)
    
    X_test.reset_index(drop=True, inplace=True)
    
    shap_values = explainer.shap_values(X_test) #, check_additivity=False)
        
    stats = pd.DataFrame(shap_values).describe()
    means = stats.loc[["mean"]]
    means = means.sort_values(by='mean', ascending=False, axis=1)

    c = X_test.iloc[:, means.columns].columns
    means = means.set_axis(c, axis=1, inplace=False)
    
    g_c = []
    a_c = []
    for c in input_concepts:
        # feature contributions g_c for output o
        g_c.append(means[c].values[0].sum())
        a_c.append(means[c].values[0])
    return(g_c, a_c, shap_values, means, X_test)
    
```
